### Monotower parameters

```python
Gs = 10e6 # soil modulus, [Pa]
```

```python
rho_s = 2000 # soil density, [kg/m^3]
```

```python
vs = 0.33 #Poisson's ratio
```

```python
k1 = lambda omega0: 2.16e9*(1-0.0318*omega0) # soil stiffness, sliding, [N/m]
```

```python
k_theta = lambda omega0: 3.63e12*(1-0.137*omega0) # soil stiffness, rocking, [N-m]
```

```python
m = 6.15e8 # structural mass, virtual, [kg]
```

```python
m0 = 3.56e8 # structural mass, actual, [kg]
```

```python
mb = 2.59e8 # structural mass, buoyant, [kg]
```

```python
JG = 8.8e11 # structural inertia, virtual, [kg-m^2]
```

```python
J0 = 1.66*JG # structural inertia, [kg-m^2]
```

```python
hG = 30.7 # center of actual mass, [m]
```

```python
hb = 31.7 # center of buyant mass, [m]
```

```python
r0 = 45 # caisson radius, [m]
```

```python
# computed frequencies, [rad/sec]
omega1 = 1.24
omega2 = 2.73
```

```python
g=9.81 # the acceleration due to gravity, [m/sec^2]
```

### Calculations

```python
M=Matrix([[m,0],
            [0,JG]])
```

```python
K=Matrix([[J0/JG*k1(1.24), -m*hG/JG*(k_theta(1.24)-m0*g*hG+mb*g*hb)],
            [-hG*k1(1.24), (k_theta(1.24)-m0*g*hG+mb*g*hb)]])
```

```python
omega_squad = symbols('omega_squad')
```

```python
solutions = solve((K-omega_squad *M).det(), omega_squad )
```

```python
solutions[0], solutions[1] 
```

#### Omega_squad is real and positive, hence the structure is stable! 

```python
omega1_ = sqrt(solutions[0])
omega2_ = sqrt(solutions[1])
```

```python
omega1_, omega2_ 
```

```python
xi = symbols('xi')
```

```python
solutions = (K-omega_squad *M)*Matrix([1,xi])
```

```python
solutions[0]
```

```python
solutions[1]
```

```python
xi21_ = solve(solutions[0], xi)[0].subs(omega_squad, omega1**2)
```

```python
xi22_ = solve(solutions[1], xi)[0].subs(omega_squad, omega2**2)
```

```python
xi21_, xi22_
```
