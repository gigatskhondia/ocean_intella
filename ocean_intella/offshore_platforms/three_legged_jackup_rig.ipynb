```python
import scipy.integrate as integrate
import numpy as np
```

### Input

```python
rho_g = 64.3 # water density, [lb/ft^3]
```

```python
D = 12 # single leg diameter, [ft]
```

```python
CM = 2.0 # assumed inertia coefficient
```

```python
m = 4.42e5 # slugs, equivalent mass [lb-sec^2/ft]
```

```python
k1 = 8.18e5 # bending stiffness, [lb/ft]
```

```python
c1 = 6.01e4 # damping, [lb-sec/ft]
```

```python
Hs = 15 # significant wave height,  [m]
```

```python
B = 3.11/Hs**2
```

```python
g = 32.2 # the acceleration due to gravity , [ft/sec^2]
```

### Functions

```python
G_omega2 = (3*np.pi/8*rho_g*D**2*CM)**2
```

```python
S_eta = lambda omega: 0.0081*g**2/omega**5*np.exp(-B/omega**4) # Pierson-Moskowitz spectrum
```

### Calculations

```python
result = integrate.quad(lambda omega: 2*G_omega2*S_eta(omega)/((k1-m*omega**2)**2+c1**2*omega**2), 0.16, 1.6)
```

```python
result
```

```python
sigma = np.sqrt(result)[0]
```

```python
sigma # standard deviation
```

```python
v = 3*sigma # deck displacement, [ft]
```

```python
v
```

```python
fmax = k1*v/3 # horizontal shear load per leg, [lb]
```

```python
fmax
```

```python
l = 265 # leg height, [ft]
```

```python
Mmax = l*fmax # overturning moment, [ft-lb]
```
